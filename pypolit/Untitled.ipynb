```python
from datetime import datetime
```

```python
def date_norm(date):
    d = datetime.fromtimestamp(date)
    str_d = d.strftime("%Y-%m-%d %H:%M:%S")
    return str_d
```

```python
df['date_norm'] = df['date'].apply(date_norm) 
```

```python
new = df['date_norm'].str.split(expand = True)  
```

```python
df['Date'] = new[0]
df['Time'] = new[1] 
```

```python
from vk_api.longpoll import VkLongPoll, VkEventType
```

```python
key = "e5b4e077cc43856fb3ba6d2a270e1e92c7b0214ef2ccb36187caba7aa5b37bcb4e1f580cedc33b6092bab"
```

```python
vk_session = vk_api.VkApi(token=key)
longpoll = VkLongPoll(vk_session) 

vk = vk_session.get_api() 
```

```python
for event in longpoll.listen():
    if event.type == VkEventType.MESSAGE_NEW and event.to_me and event.text:
        if event.text == "привет":
            if event.from_user:
                vk.messages.send(
                user_id = event.user_id, 
                message = "Привет! В Москве хмарь.",
                random_id = 123456)
        if event.text == "пока":
            if event.from_user:
                vk.messages.send(
                user_id = event.user_id, 
                message = "Au revoir",
                random_id = 12345689)
```
